```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **ALGORITMO K - MEANS**

```python
from sklearn.datasets import make_blobs
#Generar datos sintticos bidimensionales.
X, y=make_blobs(n_samples=500, centers=3, random_state=42, cluster_std=1.8)
#Representacion grafica de los datos
fig, ax=plt.subplots(figsize=(8,4))
sns.scatterplot(x=X[:,0], y=X[:,1], s=50)
```

Es evidente que hay tres clsteres en nuestros datos de muestra. A continuacin, entrenamos el modelo especificando tres clsteres y luego lo utilizamos para predecir las etiquetas de agrupacin para cada punto de datos.

**Especificar el nombre de los clusters**

```python
from sklearn.cluster import KMeans
#Especificar el nombre de los clusters
kmeans=KMeans(n_clusters=3)
#Entrenar el modelo
kmeans.fit(X)
#Predecir los clusters
y_pred=kmeans.predict(X)
print(kmeans.cluster_centers_)
print(kmeans.labels_)
```

Despus de entrenar el modelo, podemos usar los atributos del estimador cluster_centers_ para comprobar las coordenadas de los centros de los clsteres y labels_ para comprobar las etiquetas de cada punto de datos:

```python
#Vamos a trazar los resultados con los tres centros:
fig, ax=plt.subplots(figsize=(8,4))
sns.scatterplot(x=X[:,0], y=X[:,1], hue=y_pred, s=50, palette='viridis')

centers=kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], marker='^', c='red', s=80)
```

Explicacin: Este bloque se centra en la visualizacin de resultados de K-Means Clustering (un algoritmo de aprendizaje no supervisado, diferente a la Regresin Logstica de los pasos anteriores).

sns.scatterplot(...): Grafica los puntos de datos (ej. las dos primeras caractersticas de X), coloreados segn el clster asignado (y_pred).

centers=kmeans.cluster_centers_: Obtiene las coordenadas de los centros de los clsteres.

plt.scatter(...): Superpone los centros de los clsteres en el grfico, marcados con un tringulo rojo.

Uno de los retos del mtodo k-means es que debemos predeterminar el nmero de clsteres en el modelo, es decir, el algoritmo no puede aprender por s mismo a partir de los datos. Por ejemplo, si le indicamos al modelo que prediga cinco clsteres, proceder a intentar encontrar los mejores resultados:

```python
#Especificar el numero de Clusters
kmeans=KMeans(n_clusters=5, random_state=42)
#Entrenar el modelo
kmeans.fit(X)
#Predecir los clusters
y_pred=kmeans.predict(X)
#Grupo de clusters
fig, ax=plt.subplots(figsize=(8,4))
sns.scatterplot(x=X[:,0], y=X[:,1], hue=y_pred, s=50, palette='viridis')
centers=kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], marker='^', c='red', s=80)
```

Explicacin: Este bloque instancia, entrena y visualiza el modelo K-Means.

Comentario en espaol: Explica el reto de predeterminar el nmero de clsteres (k).

kmeans=KMeans(n_clusters=5, ...): Instancia el modelo K-Means, especificando que busque 5 clsteres.

kmeans.fit(X): Entrena el modelo, encontrando los 5 centros ptimos.

y_pred=kmeans.predict(X): Asigna una etiqueta de clster a cada punto de datos.

Las lneas de sns.scatterplot y plt.scatter grafican los datos y los centros de los 5 clsteres.
